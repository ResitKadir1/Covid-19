```python
import pandas as pd
from IPython.display import Javascript
from IPython.core.display import display
from IPython.core.display import HTML
```

```python
#  Datasets
df_confirmed= pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df_deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
df_recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
                           
```

```python
#lets group countries
conf_df_grouped = df_confirmed.groupby('Country/Region')[df_confirmed.columns[4:]].sum().reset_index()
deaths_df_grouped = df_deaths.groupby('Country/Region')[df_deaths.columns[4:]].sum().reset_index()
recv_df_grouped = df_recovered.groupby('Country/Region')[df_recovered.columns[4:]].sum().reset_index()

```

```python
#Get flag urls
country_flags = pd.read_csv('country_flags_url.csv')
country_flags.head(2)
```

```python
#Lets insert flags on data
deaths_df_grouped.insert(1,'Flag URL','')
conf_df_grouped.insert(1,'Flag URL','')
recv_df_grouped.insert(1,'Flag URL','')
deaths_df_grouped.head()
```

```python
flag = country_flags[country_flags['Country Name']=="Denmark"]['Image URL']
flag = list(flag)[0]
flag
```

```python
countries_infected = df_confirmed['Country/Region'].unique()
```

```python
#insert url on the Countries
flags_unk = []
for cnt in countries_infected:
    flag = country_flags[country_flags['Country Name']==cnt]['Image URL']
    flag = list(flag)
    if flag:
        index = conf_df_grouped[conf_df_grouped['Country/Region']==cnt].index
        conf_df_grouped.at[index,'Flag URL']=flag[0]
    else:
        flags_unk.append(cnt)

for cnt in countries_infected:
    flag = country_flags[country_flags['Country Name']==cnt]['Image URL']
    flag = list(flag)
    if flag:
        index = deaths_df_grouped[deaths_df_grouped['Country/Region']==cnt].index
        deaths_df_grouped.at[index,'Flag URL']=flag[0]

    
for cnt in countries_infected:
    flag = country_flags[country_flags['Country Name']==cnt]['Image URL']
    flag = list(flag)
    if flag:
        index = recv_df_grouped[recv_df_grouped['Country/Region']==cnt].index
        recv_df_grouped.at[index,'Flag URL']=flag[0]
```

```python
#SAve the updated version
conf_df_grouped.to_csv('covid_19_race_bar_conf.csv', index=False)
deaths_df_grouped.to_csv('covid_19_race_bar_death.csv', index=False)
recv_df_grouped.to_csv('covid_19_race_bar_recv.csv', index=False)
```
